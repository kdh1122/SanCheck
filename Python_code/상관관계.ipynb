```python
data_route = '/content/gdrive/MyDrive/final_data_컬럼통합m.csv'
```

```python
accident = pd.read_csv(data_route, encoding='euc-kr')
accident
```

```python
accident = accident.rename(columns={'사고원인코드명_사고종별': '사고원인'})
```

```python
accident = accident.fillna(0)
accident
```

```python
accident['개인(급.만성)질환 여부'] = accident['사고원인'].apply(lambda x: 1 if x == '개인(급.만성)질환' else 0)
accident['개인질환 여부'] = accident['사고원인'].apply(lambda x: 1 if x == '개인질환' else 0)
accident['고온환경질환 여부'] = accident['사고원인'].apply(lambda x: 1 if x == '고온환경질환' else 0)
accident['기타산악 여부'] = accident['사고원인'].apply(lambda x: 1 if x == '기타산악' else 0)
accident['낙석·낙빙 여부'] = accident['사고원인'].apply(lambda x: 1 if x == '낙석·낙빙' else 0)
accident['산악기타 여부'] = accident['사고원인'].apply(lambda x: 1 if x == '산악기타' else 0)
accident['실족추락 여부'] = accident['사고원인'].apply(lambda x: 1 if x == '실족추락' else 0)
accident['암벽등반 여부'] = accident['사고원인'].apply(lambda x: 1 if x == '암벽등반' else 0)
accident['야생식물 섭취 중독 여부'] = accident['사고원인'].apply(lambda x: 1 if x == '야생식물 섭취 중독' else 0)
accident['일반조난 여부'] = accident['사고원인'].apply(lambda x: 1 if x == '일반조난' else 0)
accident['저체온증 여부'] = accident['사고원인'].apply(lambda x: 1 if x == '저체온증' else 0)
accident['탈진·탈수 여부'] = accident['사고원인'].apply(lambda x: 1 if x == '탈진·탈수' else 0)
accident['자살기도(산악) 여부'] = accident['사고원인'].apply(lambda x: 1 if x == '자살기도(산악)' else 0)
```
