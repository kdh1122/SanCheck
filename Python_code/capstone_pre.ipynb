```python
import numpy as np
import pandas as pd

from google.colab import drive
drive.mount('/content/gdrive')
```

```python
data_route = '/content/gdrive/MyDrive/소방청_전국 산악사고 현황(2017_2021).csv'
```

```python
mountain_data = pd.read_csv(data_route, encoding='euc-kr')
mountain_data
```

```python
mountain_data['일시'] = mountain_data.apply(lambda x: str(x['신고년월일']) + ' ' + x['신고시각'], axis=1)
print(mountain_data)
```

```python
data_route1 = '/content/gdrive/MyDrive/OBS_17.csv'
data_route2 = '/content/gdrive/MyDrive/OBS_18.csv'
data_route3 = '/content/gdrive/MyDrive/OBS_19.csv'

obs17 = pd.read_csv(data_route1, encoding='euc-kr')
obs18 = pd.read_csv(data_route2, encoding='euc-kr')
obs19 = pd.read_csv(data_route3, encoding='euc-kr')
```

```python
obs_all = obs17.append([obs18, obs19], ignore_index=True)
```

```python
mountain_data['일시'] = pd.to_datetime(mountain_data['일시'])

mountain_data['report_time_simple'] = mountain_data['일시'].apply(lambda x: x.strftime('%Y-%m-%d %H:00'))
mountain_data
```

```python
obs_all
```

```python
mountain_data
```

```python
obs_all = obs_all.rename(columns={'일시': 'report_time_simple'})
```

```python
mountain_data = mountain_data.rename(columns={'지점코드': '지점코드1', '지점코드2': '지점코드'})
```

```python
result1 = pd.merge(mountain_data, obs_all, on=['report_time_simple', '지점코드'])
```

```python
result1
```

```python
result2 = pd.merge(mountain_data, obs_all, on=['report_time_simple', '지점코드'])
```

```python
accident = result1.append([result2], ignore_index=True)
accident
```
