```python
accident1_route = '/content/gdrive/MyDrive/사고_2010_2016.csv'
accident2_route = '/content/gdrive/MyDrive/사고_2017_2021.csv'
```

```python
accident_data1 = pd.read_csv(accident1_route, encoding='euc-kr')
accident_data2 = pd.read_csv(accident2_route, encoding='euc-kr')
accident_data1
accident_data2
```

```python
weather_route1 = '/content/gdrive/MyDrive/OBS_10.csv'
weather_route2 = '/content/gdrive/MyDrive/OBS_11.csv'
weather_route3 = '/content/gdrive/MyDrive/OBS_12.csv'
weather_route4 = '/content/gdrive/MyDrive/OBS_13.csv'
weather_route5 = '/content/gdrive/MyDrive/OBS_14.csv'
weather_route6 = '/content/gdrive/MyDrive/OBS_15.csv'
weather_route7 = '/content/gdrive/MyDrive/OBS_16.csv'
weather_route8 = '/content/gdrive/MyDrive/OBS_17.csv'
weather_route9 = '/content/gdrive/MyDrive/OBS_18.csv'
weather_route10 = '/content/gdrive/MyDrive/OBS_19.csv'
weather_route11 = '/content/gdrive/MyDrive/OBS_20.csv'
weather_route12 = '/content/gdrive/MyDrive/OBS_21.csv'


obs10 = pd.read_csv(weather_route1, encoding='euc-kr')
obs11 = pd.read_csv(weather_route2, encoding='euc-kr')
obs12 = pd.read_csv(weather_route3, encoding='euc-kr')
obs13 = pd.read_csv(weather_route4, encoding='euc-kr')
obs14 = pd.read_csv(weather_route5, encoding='euc-kr')
obs15 = pd.read_csv(weather_route6, encoding='euc-kr')
obs16 = pd.read_csv(weather_route7, encoding='euc-kr')
obs17 = pd.read_csv(weather_route8, encoding='euc-kr')
obs18 = pd.read_csv(weather_route9, encoding='euc-kr')
obs19 = pd.read_csv(weather_route10, encoding='euc-kr')
obs20 = pd.read_csv(weather_route11, encoding='euc-kr')
obs21 = pd.read_csv(weather_route12, encoding='euc-kr')
```

```python
obs_all = obs10.append([obs11, obs12, obs13, obs14, obs15, obs16, obs17, obs18, obs19, obs20, obs21], ignore_index=True)
```

```python
obs_all
```

```python
obs_all = obs_all.rename(columns={'일시': 'report_time_simple'})
```

```python
accident = accident_data1.append([accident_data2], ignore_index=True)
```

```python
accident
```

```python
accident1 = accident.rename(columns={'지점코드1': '지점코드'})
result1 = pd.merge(accident1, obs_all, on=['report_time_simple', '지점코드'])
```

```python
result1
```

```python
accident2 = accident.rename(columns={'지점코드2': '지점코드'})
result2 = pd.merge(accident2, obs_all, on=['report_time_simple', '지점코드'])
```

```python
result2
```

```python
accident = result1.append([result2], ignore_index=True)
accident
```

```python
accident = accident.drop('지점코드1', axis=1)
accident = accident.drop('지점코드2', axis=1)
accident
```

```python
accident = accident.fillna(0)
accident
```

```python
accident.to_csv('/content/gdrive/MyDrive/final_data.csv', index=False)
```
